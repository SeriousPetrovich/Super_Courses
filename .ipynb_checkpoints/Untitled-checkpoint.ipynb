```python
import matplotlib.pyplot as plt
import random
import math

k = 1001

def x_k(input):
    return 3*input/k

def func_x(input):
    x = x_k(input)
    return 1/(x**2 - 3*x +2)

def func_y(k):
    func = func_x(k)
    if func < -100:
        result = -100 + random.random()
    elif func > 100:
        result = -100 + random.random()
    else:
        result = func + random.random()
    return result
```

```python
plt.plot(list(x_k(i) for i in range(0,k)), list(func_y(i) for i in range(0,k)))
plt.show()
```

```python
def func_fit(x,input2):
    a = input2[0]
    b = input2[1]
    c = input2[2]
    d = input2[3]
    return (a*x + b)/(x**2 + c*x + d)

def func_d(input2):
    
    return sum( (func_fit(x_k(i), input2) - func_y(i))**2  for i in range(0,k))
```

```python
import numpy as np
from scipy.optimize import minimize
```

```python
x0 = np.array([1, -1, -1, 1])
res = minimize(func_d, [1,-1,-1,1],method='nelder-mead',options={'disp':True,'fatol':0.01})
```

```python
print(res.x)
```

```python
ala = list(x_k(i) for i in range(0,k))
olo =  list(func_fit(x_k(i),res.x) for i in range(0,k))
```

```python
plt.plot(ala, olo, 'red')
plt.plot(list(x_k(i) for i in range(0,k)), list(func_y(i)for i in range(0,k)))
plt.show()
```

```python
plt.plot(list(x_k(i) for i in range(0,k)), list(func_fit(x_k(i),res.x) for i in range(0,k)), 'red')
plt.plot(list(x_k(i) for i in range(0,k)), list(func_y(i)for i in range(0,k)))
plt.show()
```
